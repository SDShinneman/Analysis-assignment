My story idea is to compare racially charged offenses on campuses in Missouri to those offenses on campuses across the country at large. The story would take a look at how Missouri stacks up in crime areas prone to racial bias, such as intimidation and vandalism. In those areas, eight intimidations and nine vandalisms were recorded in Missouri from 2011 through 2013. That compares to 363 intimidations and 510 vandalisms on US campuses during that time.

That means Missouri's portion comes out to 2.2 percent and 1.7 percent -- not very sexy numbers, as Missouri, obviously, represents 2 percent of the total states in the U.S. Some population analysis would need to be done to see what portion of total U.S. college students are in Missouri. But showing how Missouri stacks up is important regardless of whether they are above average or below. With so much of the country's current racial tension taking its roots from here in Missouri, this story would provide a little context as to the actual level of racially charged incidents that occur here.
